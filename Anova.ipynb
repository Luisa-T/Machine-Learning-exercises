```python
# numpy is used to create random variables
import numpy as np
# pandas is used to handle the data
import pandas as pd
# matplotlib is used to create the plots
import matplotlib.pyplot as plt
# scipy is used to create the t-tests and ANOVA
import scipy.stats as stats
```

```python
#setting the highest point of the bell shaped curve (m) for each of the populations
popA = {'m': 1.6, 's': 0.1}
popB = {'m': 1.8, 's': 0.1} # s is the standard deviation

#creating random data for each of the above populations
sampA = np.random.normal(popA['m'], popA['s'], 100)
sampB = np.random.normal(popB['m'], popB['s'], 100)
```

```python
# Calculate for each sample, the independen samples t-statisitc and the pobability that other samples are as different as those two if everything else is equal
stats.ttest_ind(sampA, sampB)
```

### Problem with t-tests

#### Multiple t-tests and Type I error

Think of coin flips.
For t-tests we test whether the null hypothesis is true which means the propability of obtaining a not significant result which means 1 - 0.05 = 0.95
***
So the more t-tests you do you multiply the 0.95 by the number of tests.
***
The result is then subtracted from 1 and the result of that is then the pobability of making at least one type I error.
Essentially, the more inferences you make, the more likely it is to make an erroneous inference.
***
ANOVA compares all means at the same time and maintains the type I error probability at the initial level.

#### About t-tests

When we perform an independent sample t-test we assume that there is a difference between the means of the two population we are testing.
So we are asking the test to tell us what is the likelihood to see a t-statistic as extreme as the current sample.

***
#### Errors
There is always a chance that we will see a t-statistic as extreme as the current sample.

We also may reect the hypothesis incorrectly, meaning that the populations may in fact be statistically different but the current sample may just randomly may not show this difference.

Type I and Type II images from https://en.wikipedia.org/wiki/Type_I_and_type_II_errors

#### Manually calculating the t-statistic

We are using two samples, with equal variances, and possibly different sample sizes

$$ t=X¯1−X¯2sp⋅1n1+1n2−−−−−−−√
sp=(n1−1)s2X1+(n2−1)s2X2n1+n2−2−−−−−−−−−−−−−−−−−−−−−−√
s2=1n−1∑i=1n(Yi−Y¯¯¯¯)2 $$

```python
# Counting the sample
nA = float(len(sampA))
nB = float(len(sampB))

# Calulate the means for each sample
mA = sampA.sum() / nA
mB = sampB.sum() / nB

# calculate the variance in the samples
varA = ((sampA - mA)**2).sum() / (nA - 1.0)
varB = ((sampB - mB)**2).sum() / (nB -1.0)

# Standard deviation in the pool
sp = np.sqrt(((nA - 1.0)  * varA + (nB - 1.0) * varB) / (nA + nB - 2.0))

# calculate the t-statistic
t = (mA - mB) / (sp * np.sqrt((nB - 1.0 ) * varB) / (nA + nB - 2.0))

print(f"The mean of sample A is: {mA:8.4f}")
print(f"The mean of sample B is: {mB:8.4f}")
print(f"The size of sample A is: {nA:8.4f}")
print(f"The size of sample B is: {nB:8.4f}")
print(f"The variance of sample A is: {varA:8.4f}")
print(f"The variance of sample A is: {varB:8.4f}")
print(f"The pooled sandard deviation is: {sp:8.4f}")
print(f"The t-statistic is: {t:8.4f}")
```

```python
# x values for plotting
x = np.linspace(-5.0, 5.0, 1000)

# The pdf for the t-distribution
# The number of degrees of freedom s (nA + nB -2)
pdf = stats.t.pdf(x, (nA + nB - 2.0))

# Use x and pdf to create a dataframe
df = pd.DataFrame({'x': x, 'y': pdf})

# Plot the overall distribution
plt.fill_between(df['x'], df['y'], color="r", alpha=0.25)

# Plot the values which are more extreme than our |t|
crit = np.abs(stats.t.ppf(0.975, nA + nB - 2.0))
tail1 = df[df['x'] >= crit]
tail2 = df[df['x'] <= - crit]
plt.fill_between(tail1['x'], tail1['y'], color='g', alpha=0.25)
plt.fill_between(tail2['x'], tail2['y'], color='g', alpha=0.25)

print(crit)

plt.show()
```
