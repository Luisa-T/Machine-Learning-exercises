```python
import pandas as pd
import sklearn.preprocessing as pre
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
df
```

```python
x = df.iloc[:, 0:4]
x
```

```python
# https://scikit-learn.org/stable/modules/preprocessing.html
pre.scale(x)
```

```python
xscale = pd.DataFrame(pre.scale(x), columns=x.columns)
xscale
```

```python
(x - x.mean()) / x.std()
```

```python
xscale.mean()
```

```python
xscale.std()
```

```python
xscale.var()
```

```python
scaler = pre.StandardScaler()
scaler.fit(x)
scaler
scaler.mean_
scaler.scale_
```

```python
x1scale = pd.DataFrame(scaler.transform(x), columns = x.columns)
```

```python
scaler.transform(np.array([[1.0, 0.5, 10.0, 4.1]]))
```

### whitening of the data

```python
x.corr()
```

```python
import sklearn.decomposition as dec
```

```python
pca = dec.PCA(n_components=4, whiten=True)
pca.fit(x)
xwhite = pd.DataFrame(pca.transform(x), columns=x.columns)
xwhite
```

```python
xwhite.corr().round()
```

```python
xwhite.mean().round()
```

```python
xwhite.std().round()
```
