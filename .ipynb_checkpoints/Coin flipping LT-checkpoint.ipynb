# Coin flipping

***

Coin flipping and the ideas behind it are important to computing.
Computers rely on bits which are very similar to the two sides of a coin, one and zero, heads and tails.

Flipping a coin therefore has exactly two outcomes.

```python
# numpy and scpy.stats are used to analyse flipping a coin
import numpy as np
import scipy.stats as ss

# to visualise we use matplotlib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# to increase the default plot size
plt.rcParams['figure.figsize'] = (12,8)
```

A fair coin will give each outcome an equal chance, 50% heads, and 50% tails or translated into probability 0.5 each.
The below will simulate this where 1 equals heads and 0 equals tails.

```python
# the number of times to flip a coin, the probability to get heads, the number of times to do so
np.random.binomial(1, 0.5, 1)
```

```python
# flipping a coin 1000 times, how likely is it to get heads?
np.random.binomial(1000, 0.5, 1)
```

***
How likely is it to see a specific number of heads when flipping a coin x amount of times?
***

```python
# the number of heads, the number of times to flip the coin, the probability to get heads
ss.binom.pmf(521, 1000, 0.5)
```

```python
# graphic representation to get a certain outcome
sns.distplot(np.random.binomial(1000, 0.5, 1000))
```

***
Supposing we have an unfair coin
***

```python
# how man heads do we get when we flip an unfair coin?
np.random.binomial(10, 0.2, 1)
```

***
The probability of r heads when flipping an unfair coin n times is

$$ p(r∣n,p)=(nr)pr(1−p)(n−r) $$

***

```python
noflips = 10

p = 0.3

d = [ss.binom.pmf(i, noflips, p) for i in range(noflips + 1)]
d
```

***

n over r is the number of ways to select r items from n, ignoring the order you select them in

***

```python
import math

n = 10
r = 6

choose = lambda x, y: math.factorial(x) / (math.factorial(y) * math.factorial(x-y))

choose(n, r)
```

***

the following applies for n over 0 and n over n

***

```python
choose(10, 0)
```
